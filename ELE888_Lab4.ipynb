```python
#Q2

import numpy as np

# Function to generate x and y values
def generate_data(num_samples=30000):
    x_values = np.random.uniform(-1, 1, num_samples)
    y_values = 0.2 * x_values**4 + 2 * x_values**3 + 0.1 * x_values**2 + 10
    return x_values, y_values

# Function to shuffle data
def shuffle_data(x_values, y_values):
    indices = np.random.permutation(len(x_values))
    return x_values[indices], y_values[indices]

# Function to split data into train, validation, and test sets
def split_data(x_values, y_values, train_ratio=0.7, val_ratio=0.15, test_ratio=0.15):
    assert train_ratio + val_ratio + test_ratio == 1.0, "Ratios must sum to 1.0"

    num_samples = len(x_values)
    num_train = int(train_ratio * num_samples)
    num_val = int(val_ratio * num_samples)

    x_train, y_train = x_values[:num_train], y_values[:num_train]
    x_val, y_val = x_values[num_train:num_train+num_val], y_values[num_train:num_train+num_val]
    x_test, y_test = x_values[num_train+num_val:], y_values[num_train+num_val:]

    return (x_train, y_train), (x_val, y_val), (x_test, y_test)

# Function to get dataset with optional shuffling
def get_dataset(shuffle=False):
    x_values, y_values = generate_data()
    if shuffle:
        x_values, y_values = shuffle_data(x_values, y_values)
    return x_values, y_values

# Example usage:
# Without shuffling
x_values, y_values = get_dataset(shuffle=False)
print("Without shuffling:")
print("x_values:", x_values[:5])
print("y_values:", y_values[:5])

# With shuffling
x_values_shuffled, y_values_shuffled = get_dataset(shuffle=True)
print("\nWith shuffling:")
print("x_values:", x_values_shuffled[:5])
print("y_values:", y_values_shuffled[:5])

# Splitting data
(x_train, y_train), (x_val, y_val), (x_test, y_test) = split_data(x_values, y_values)
print("\nTrain set size:", len(x_train))
print("Validation set size:", len(x_val))
print("Test set size:", len(x_test))
```

```python
#Q4

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def evaluate_predictions(actual, predicted):
    # Calculate Mean Absolute Error (MAE)
    mae = mean_absolute_error(actual, predicted)

    # Calculate Mean Squared Error (MSE)
    mse = mean_squared_error(actual, predicted)

    # Calculate Root Mean Squared Error (RMSE)
    rmse = np.sqrt(mse)

    # Calculate R^2 Score
    r2 = r2_score(actual, predicted)

    return mae, mse, rmse, r2

# Example usage:
actual = [3, -0.5, 2, 7]
predicted = [2.5, 0.0, 2, 8]
mae, mse, rmse, r2 = evaluate_predictions(actual, predicted)

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R^2 Score:", r2)
```

```python
#Q5

import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Function to generate x and y values
def generate_data(num_samples=30000):
    x_values = np.random.uniform(-1, 1, num_samples)
    y_values = 0.2 * x_values**4 + 2 * x_values**3 + 0.1 * x_values**2 + 10
    return x_values, y_values

# Function to shuffle data
def shuffle_data(x_values, y_values):
    indices = np.random.permutation(len(x_values))
    return x_values[indices], y_values[indices]

# Function to split data into train, validation, and test sets
def split_data(x_values, y_values, train_ratio=0.3, val_ratio=0.2, test_ratio=0.5):
    x_train, x_temp, y_train, y_temp = train_test_split(x_values, y_values, test_size=(1 - train_ratio), random_state=42)
    x_val, x_test, y_val, y_test = train_test_split(x_temp, y_temp, test_size=test_ratio / (test_ratio + val_ratio), random_state=42)
    return (x_train, y_train), (x_val, y_val), (x_test, y_test)

# Function to scale data between 0 and 1 using Min-Max scaling
def scale_data(x_values, y_values):
    x_scaled = (x_values - np.min(x_values)) / (np.max(x_values) - np.min(x_values))
    y_scaled = (y_values - np.min(y_values)) / (np.max(y_values) - np.min(y_values))
    return x_scaled, y_scaled

# Function to build neural network model
def build_model(structure, activation):
    model = Sequential()
    for units in structure:
        model.add(Dense(units, activation=activation))
    model.add(Dense(1))  # Output layer
    model.compile(loss='mse', optimizer='adam')
    return model

# Function to evaluate model performance
def evaluate_model(model, x_test, y_test):
    y_pred = model.predict(x_test)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    return mae, mse, rmse, r2, y_pred

# Function to plot actual vs predicted values
def plot_results(x_test, y_test, y_pred):
    plt.figure(figsize=(10, 6))
    plt.scatter(x_test, y_test, color='blue', label='Actual')
    plt.scatter(x_test, y_pred, color='red', label='Predicted')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Actual vs Predicted')
    plt.legend()
    plt.grid(True)
    plt.show()

# Function to run a specific case
def run_case(structure, activation, shuffled, scaled):
    # Generate data
    x_values, y_values = generate_data()
    if shuffled:
        x_values, y_values = shuffle_data(x_values, y_values)
    (x_train, y_train), (x_val, y_val), (x_test, y_test) = split_data(x_values, y_values)
    if scaled:
        x_train, y_train = scale_data(x_train, y_train)
        x_val, y_val = scale_data(x_val, y_val)
        x_test, y_test = scale_data(x_test, y_test)

    # Reshape input data
    x_train = x_train.reshape(-1, 1)  # Reshape to (batch_size, input_dim)
    x_val = x_val.reshape(-1, 1)
    x_test = x_test.reshape(-1, 1)

    # Build and train model
    model = build_model(structure, activation)
    model.fit(x_train, y_train, epochs=20, batch_size=12, verbose=0, validation_data=(x_val, y_val))

    # Evaluate model
    mae, mse, rmse, r2, y_pred = evaluate_model(model, x_test, y_test)

    # Plot results
    plot_results(x_test, y_test, y_pred)

    return mae, mse, rmse, r2

# Define parameters
structure_1 = [12, 8, 4]
structure_2 = [24]
activations = ['relu', 'tanh']
shuffled_options = [True, False]
scaled_options = [True, False]

# Run all cases
for activation in activations:
    for shuffled in shuffled_options:
        for scaled in scaled_options:
            print(f"\nActivation: {activation}, Shuffled: {shuffled}, Scaled: {scaled}")
            if shuffled and scaled:
                print("Data is shuffled and scaled.")
            elif shuffled:
                print("Data is shuffled.")
            elif scaled:
                print("Data is scaled.")
            else:
                print("Data is neither shuffled nor scaled.")
            mae, mse, rmse, r2 = run_case(structure_1, activation, shuffled, scaled)
            print("MAE:", mae)
            print("MSE:", mse)
            print("RMSE:", rmse)
            print("R2 Score:", r2)
```

```python
#Q6

import numpy as np

# Define the sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define the derivative of the sigmoid function
def sigmoid_derivative(x):
    return x * (1 - x)

# Define the neural network class
class NeuralNetwork:
    def __init__(self):
        # Seed the random number generator for reproducibility
        np.random.seed(42)

        # Initialize weights
        self.weights_input_hidden = np.random.rand(2, 2)  # Weights for input to hidden layer
        self.weights_hidden_output = np.random.rand(2, 1)  # Weights for hidden to output layer

        # Initialize biases
        self.bias_hidden = np.random.rand(1, 2)  # Bias for hidden layer
        self.bias_output = np.random.rand(1, 1)  # Bias for output layer

    def feedforward(self, X):
        # Input to hidden layer
        self.hidden_output = sigmoid(np.dot(X, self.weights_input_hidden) + self.bias_hidden)

        # Hidden to output layer
        self.output = sigmoid(np.dot(self.hidden_output, self.weights_hidden_output) + self.bias_output)
        return self.output

    def backpropagation(self, X, y, learning_rate):
        # Calculate the error
        error = y - self.output

        # Calculate gradients for the output layer
        output_gradients = error * sigmoid_derivative(self.output)

        # Calculate gradients for the hidden layer
        hidden_gradients = np.dot(output_gradients, self.weights_hidden_output.T) * sigmoid_derivative(self.hidden_output)

        # Update weights and biases
        self.weights_hidden_output += np.dot(self.hidden_output.T, output_gradients) * learning_rate
        self.weights_input_hidden += np.dot(X.T, hidden_gradients) * learning_rate
        self.bias_output += np.sum(output_gradients, axis=0) * learning_rate
        self.bias_hidden += np.sum(hidden_gradients, axis=0) * learning_rate

    def train(self, X, y, epochs=1000, learning_rate=0.1):
        for epoch in range(epochs):
            output = self.feedforward(X)
            self.backpropagation(X, y, learning_rate)
            if epoch % 100 == 0:
                print(f'Epoch {epoch}: Loss = {np.mean(np.abs(y - output))}')

# Define the XOR truth table
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

y = np.array([[0],
              [1],
              [1],
              [0]])

# Create an instance of the neural network
model = NeuralNetwork()

# Train the neural network
model.train(X, y, epochs=10000, learning_rate=0.1)

# Evaluate the trained model
predictions = model.feedforward(X)
print("Predictions: ")
print(predictions)
```
